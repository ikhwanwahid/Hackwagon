### Question 1

Print out all of the NRIC in the given data.

<b>Expected Output:</b>

    S93250888

```python
data = [
    {'name': 'kt', 'nric': 'S93250888', 'score': 10, 'occupation': 'student'},
    {'name': 'john', 'nric': 'S93250818', 'score': 5, 'occupation': 'student'},
    {'name': 'henry', 'nric': 'S93250288', 'score': 2, 'occupation': 'teacher'},
    {'name': 'mark', 'nric': 'S93250218', 'score': 7, 'occupation': 'student'},
    {'name': 'mary', 'nric': 'S93250248', 'score': 2, 'occupation': 'teacher'},
]

#When iterating through a list of dictionaries, try to name the variable as row so that you know
#you will need to retrieve the column values out by using it's keys (in excel it will be headers)
for row in data:
    
    #retrieve the value you need using the respective keyes (headers in excel) 
    nric = row['nric'] 
    
    #lastly print the nric
    print(nric)
```

### Question 4

<h3> Find all factors of a number </h3>

<p> Given a number N, print out all numbers that are factors of N<br>
    <i> Doesn't matter if you print a list or a string </i>
</p>

**Expected output:**
    
    [1, 3, 5, 11, 15, 33, 55, 165]

```python
N = 165

#First create a list (usually blank). However in this question, we know that one will always be in
#As factors are always one and the number itself, so you don't have to use for loop to know if 
#1 is a factor or not
result = [1] 

#factors of N will never be higher than N//2, // is floor division.
#so we don't have to iterate through anything more than half of N.
for i in range(2,N//2): #remember this is a for range loop so we name the variable as i
    
    #Noticed that. I'm using N (165) divide by i 
    #for example, in the first iteration it will be 165 / 2. And you will get 82.5
    #I will then use % (modulo) which returns you the remainder of the number
    #so when I put %1 it means, i'm finding the remainder when the number is divided by 1
    #which will gives me all the decimals points. 
    #coming back to the current example, I would put 82.5 % 1 which will give me 0.5
    #I then check the remainder if its == 0. If it is true, it would mean that the N/1
    #is an integer and it will also prove that the i is a factor of N
    #Back to the example, 0.5 is not == 0. So 2 is NOT a factor of 165.
    if N/i%1 == 0:
        
        #Add numbers that are factors into our final result
        result.append(i)
        
#As factors are always one and the number itself, so you don't have to use for loop to know
#that N is also part of the factor
result.append(N) 

#Finally, print the result out
print(result)
```

### Question 5

<h3> Finding prime numbers </h3>

<p> Given a list of integers, print out all numbers that are prime numbers<br>
    <i> Doesn't matter if you print a list or a string </i>
</p>

**Expected Answers:**

    13
    31
    7
    5
    3

```python
ints = [13, 88, 21, 31, 7, 5, 1, 3, -55, 0]

#since prime number means that it is a number that can only be divided by 1 and itself
#we will first assume every element inside ints are prime numbers.
#and once we realised that the element is divisable by another number that is not 1 or itself
#we will proved that that number is NOT a prime number.
#however, if we did not managed to find any other number, that is a factor of the element
#our first assumption will prove that it is true.

#factors of N will never be higher than N//2, // is floor division.
#so we don't have to iterate through anything more than half of N.
for num in ints: #remember this is a for each loop so don't name it as i
    
    #Assuming that num is a prime number first
    #notice that prime = True is within the first for loop
    #This means for every element inside ints, I will first assume that they are True
    #You cannot place this line outside of the for loop
    prime = True
    
    #Iterate through from 2 to half of the current num
    #For example, if we take the first element of ints
    #i will be from 2 to 6 (since 6 is half of 13)
    for i in range(2,num//2): #remember this is a for range loop so we name the variable as i
        
        #Noticed that. I'm using num (13) divide by i 
        #for example, in the first iteration it will be 13 / 2. And you will get 6.5
        #I will then use % (modulo) which returns you the remainder of the number
        #so when I put %1 it means, i'm finding the remainder when the number is divided by 1
        #which will gives me all the decimals points. 
        #coming back to the current example, I would put 6.5 % 1 which will give me 0.5
        #I then check the remainder if its == 0. If it is true, it would mean that the N/1
        #is an integer and it will also prove that the i is a factor of N
        #Back to the example, 0.5 is not == 0. So 2 is also NOT a factor of 13.
        if num/i%1 == 0:
            
            #at any point when you found out that num has a factor that is not 1 or itself
            #change our assumption from True to False
            prime = False
        
    #notice that this line below is the same line as prime = True. Which means, this line will
    #run after the second for loop and at the end of every element in list.
    #if we were able to find any factors, prime would be changed in our second for loop before this
    #we also check if number is a positive integer
    if prime and num > 0:
        
        #we will only print numbers that are prime
        #notice that this is not placed at the bottom left. Because this question did not ask us
        #to keep the prime numbers in a list and print at the end
        print(num)
```

### Question 6 - Difficult

<h3> Finding number of squares from 1 to a given number, n </h3>

<p> Given a list of numbers, N. Count from 1 to N, how many perfect squares are within the number 1 to N<br>
    <i> Perfect Squares is the product of some integer with itself, meaning 4 is a perfect square as it is a product of 2x2. And 9 is a perfect square as it is a product of 3x3. </i>
</p>



**Example:**

    N = 1 -> From 1 to 1, there are 1 as perfect square -> So print 1
    N = 2 -> From 1 to 2, there are 1 as perfect square -> So print 1
    N = 3 -> From 1 to 3, there are 1 as perfect square -> So print 1
    N = 4 -> From 1 to 4, there are 1,4 as perfect square -> So print 2
    N = 5 -> From 1 to 5, there are 1,4 as perfect square -> So print 2
    N = 10 -> From 1 to 10, there are 1,4,9 as perfect square -> So print 3
    N = 26 -> From 1 to 25, there are 1,4,9,16,25 as perfect square -> So print 5
    
**Expected Output:**

    1
    2
    5
    7
    10
    

```python
list_of_numbers = [1,4,26,50,100]
#Write your code below

#iterate through the list of numbers for each number
for num in list_of_numbers:
    
    # **0.5 means square root of the num. 
    # when you square root num, it actually returns you the total number of squares from 1 to num with some deciminal places.
    # to remove the deciminal numbers, simple use // 1 (floor division) to remove all deciminal numbers
    # lastly, make sure you convert the number back to integer to remove the .0 at the back of the numbers
    print( int (num ** 0.5 // 1) )
```

### Question 7 - Easy

<h3> Palindrome Part I </h3>

<p> Given a list of words. Check if each word in the following list is a Palindrome. Prints True for those words that is a Palindrome and False for those that are not<br>
    <i> A palidrome is a string that when reverse it is still the same.</i>
</p>


**Example:**

    radar - radar - True
    refer - refer - True
    yes - sey - False
    true - eurt - False
    stats - stats - True
    level - level - True
    kaiyang - gnayiak - False
    

**Expected Output:**

    True
    False
    True
    False

```python
word_list = ["civic", "class", "rotor", "kaiyang"]
#Write code below

#iterate through the list of words for each word
for word in word_list:
    
    #you can print conditions, they will return you True / False
    #== is an equality sign to check if the string/number is the same as the other string/number
    #word[::-1] is what we learnt in string slicing where you can flip a word around
    # word == word[::-1] means to check if the word is the same when it's spelled backwards (thus, a palindrome)
    print (word == word[::-1])
```
