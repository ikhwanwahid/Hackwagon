### Question 1

Print out the names of all students in this dataset

<b>Expected Output:</b>

    KT
    John
    Henry
    Mark

```python
data = {
    'results': {
        'name': 'Classes',
        'data': [
            {
                'Class': 'A',
                'Students': ['KT', 'John']
            },
            {
                'Class': 'A',
                'Students': ['Henry', 'Mark']
            }
        ]
    }
}

#Slowly break down each of the levels (brackets)
results = data['results']
info = results['data']

#once you reached to something you know, either a list of dictionaries, dictionaries of dictionaries
#then you can use the for loop that you know
for row in info:
    students = row["Students"]
    
    #since students is another list, we can use another for loop #not advise in normal labs
    #for each student in students
    for student in students:
        
        #lastly print each student
        print(student)
```

### Question 2

Print the total score of teachers and students in data, broken down by occupation

<b>Expected Output:</b>


    {
        'teacher': 4,
        'student': 15
    }

```python
data = {
    'kt': { 'nric': 'S93250888', 'score': 10, 'occupation': 'student'},
    'john': {'nric': 'S93250818', 'score': 5, 'occupation': 'student'},
    'henry': { 'nric': 'S93250288', 'score': 2, 'occupation': 'teacher'},
    'mark': { 'nric': 'S93250218', 'score': 7, 'occupation': 'student'},
    'mark': { 'nric': 'S93250248', 'score': 2, 'occupation': 'teacher'},
}

#This is count dictionary that I have taught in class refer to the telegram chat if you don't get
#First, create a blank score dictionary
scoreDictionary = {}

#iterate through the dictionary of dictionaries
for name, row in data.items():
    
    #retrieve the specific value you need
    occupation = row['occupation']
    score = row['score']

    #check if occupation (elephant) exist within countDictionary (fridge)
    if occupation in scoreDictionary:
        
        #if exist, we do joke 2. put the giraffe in by taking elephant out
        #we were tasked to add the score instead of count number of teachers/students
        scoreDictionary[occupation] += score
        
    else:
        
        #if doesn't exist, we do joke 1. put the elephant in
        #we were tasked to add the score instead of count number of teachers/students
        scoreDictionary[occupation] = score
        
print(scoreDictionary)

#The reason why student is 15 and not 22 is because there are 2 'mark's as key in the dictionary.
#As we learnt in week 2. If there's a second key, the second value will replace the first value.
#Therefore, the first mark is not counted.
```

### Question 3

Print the average score of teachers and students in data, broken down by occupation

<b>Expected Output:</b>

    {
        'student': 7.5, 
        'teacher': 2.0
    }

```python
data = {
    'kt': { 'nric': 'S93250888', 'score': 10, 'occupation': 'student'},
    'john': {'nric': 'S93250818', 'score': 5, 'occupation': 'student'},
    'henry': { 'nric': 'S93250288', 'score': 2, 'occupation': 'teacher'},
    'mark': { 'nric': 'S93250218', 'score': 7, 'occupation': 'student'},
    'mark': { 'nric': 'S93250248', 'score': 2, 'occupation': 'teacher'},
}

#This is count dictionary that I have taught in class refer to the telegram chat if you don't get
#First, create a blank count and score dictionary, 
#like how we always create total and count for average questions
countDictionary = {}
scoreDictionary = {}

#iterate through the dictionary of dictionaries
for name, row in data.items():
    
    #retrieve the specific value you need
    occupation = row['occupation']
    score = row['score']

    #check if occupation (elephant) exist within countDictionary (fridge)
    if occupation in countDictionary:
        
        #if exist, we do joke 2. put the giraffe in by taking elephant out
        #we were tasked to derive the average
        #so we need to keep track of both total score and count of each key
        scoreDictionary[occupation] += score
        countDictionary[occupation] += 1
        
    else:
        
        #if doesn't exist, we do joke 1. put the elephant in
        #we were tasked to derive the average
        #so we need to keep track of both total score and count of each key
        scoreDictionary[occupation] = score
        countDictionary[occupation] = 1
        
        
#Next we need to find average. Average is usually total/count right? 
#This time round, the total and count are inside a dictionary. so we need to iterate through them
#But first, we must remember to create a new dictionary to store them again
averageDictionary = {}

#choose to iterate EITHER of the dictionary
for key, value in scoreDictionary.items():
    
    #notice that the key are EXACTLY the same, means we can use one key to open the other dictionary
    count = countDictionary[key]
    #score is already in the value variable that was defined 
    total_score = value
    
    #now you got your total score and count you can calculate your average
    average = total_score/count
    
    #Place the average inside each specific key component
    #We are not going to do the countDictionary concept as this is not anymore of counting
    averageDictionary[key] = average
        
#Lastly, remember to print your dictionary
print(averageDictionary)

#The reason why student is 15 and not 22 is because there are 2 'mark's as key in the dictionary.
#As we learnt in week 2. If there's a second key, the second value will replace the first value.
#Therefore, the first mark is not counted.
```

### Question 4

<h3> Largest palindrome product </h3>

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

Find the largest palindrome made from the product of two 3-digit numbers.

**Expected Answers:**

    906609

```python
#This is question is asking us to multiply two 3-digits numbers together.
#For example
# 111 * 112
# 999 * 999
# 500 * 400

#And this question also wants the result of the product of those two numbers to be a palindrome. 
#A palindrome would mean it is read exactly the same when it is flipped over

#For example,

# number  flipped over  is it a palindrome?
# 1221        1221               Yes 
# 333         333                Yes
# 456         654                No
# 764         467                No
# 848         848                Yes

#To check if its a palindrome, simply turn the number to a string and use [::-1] to flip it over
#And check for equality

#There are a few ways of doing this. The simple way, would be to iterate through 100 to 999
#Store all products that are palindrome then after afterwards sort them by the highest
#And get the highest out of them

#Create a blank list to store all product that are palindrome
palindrome_product_list = []

#Iterate through the range 100 to 1000 (exclusive) twice to get both numbers
for first_number in range(100,1000):
    for second_number in range(100,1000):
        
        #multiply first number with second number
        product = first_number * second_number
        
        #make the product into a string so that we can use [::-1]
        product_string = str(product)
        
        #Flip the number over and check for equality
        #If this is TRUE, it means that it is a palindrome
        if product_string == product_string[::-1]:
            
            #If this is TRUE. We keep the product
            palindrome_product_list.append(product)

#Sort the list from the highest to the smallest
palindrome_product_list.sort(reverse=True)

print(palindrome_product_list[0])
```

### Question 4

<h3> Largest palindrome product (a more efficient answer) </h3>

Read the following ONLY if you can understand fully the code above. This is a more efficient way of finding the same answer. Using lesser iterations and lesser function to achieve the same result.

```python
#This another way yet MORE EFFICIENT way to find 906609.

#First creating a list of numbers that only consist of 900 to 999. 
#Since the question wants the biggest palindrome, I'm guessing that the biggest product can only
#comes from the 900-999 range
number_list = list(range(900,1000))
#Flip it over so that I can test from the highest to the lowest
#This is so that I can faster find the highest
number_list = number_list[::-1]

#Create a result container to store final result
result = 0

#Create a boolean container to notice me when I first found a product that is a palindrome
break_loop = False

#Iterate through the number list get the first number and second number to multiply
for first_number in number_list:
    for second_number in number_list:
        
        #multiply first number with second number
        product = first_number * second_number
        
        #make the product into a string so that we can use [::-1]
        product_string = str(product)
        
        #Flip the number over and check for equality
        #If this is TRUE, it means that it is a palindrome
        if product_string == product_string[::-1]:
            
            #If this is TRUE. We keep the product
            result = product
            
            #When we FIRST FOUND the product that is a palindrome
            #We need STOP the for loop from continue iterating.
            #So I set by break loop equals to True to break the outer loop later
            break_loop = True
            
            #break function allows you to break out of the current for loop
            #So this break will only break the inner for loop
            break
    
    
    if break_loop:
        #while this break will break the outer for loop to ensure we get to the end
        break
            
print(result)


#Reason why I did not allow the for loop to run through all numbers and create a list to store
#all products that are palindrome and then sort it at the end is because, you don't actually need 
#the rest of the numbers in the list. You just want to find out the highest
#to find the highest, using math knowledge, you know that it is likely a number that is multiplied
#using 900++ by 900+. So there's no point in iterating through 100-899.
#Secondly, by starting from 999 down to 900. It increases your chances of finding the biggest
#in the first try. Thus, using the concept of break it allows you to cut the iterations 
#once you found the product that is palindrome.

#This code can ONLY work when you are certain that the condition met correct and there will not be
#any corner cases. Becareful when you try something like this, you must know what you are doing.
```

### Question 5

<h3> Print a Christmas tree </h3>

For a given n, where n is odd, print out a christmas tree that looks like this

When **n**  = 3

&nbsp;&nbsp;\*<br>
&nbsp;\*\*\* <br>
\*\*\*\*\* <br>

When **n**  = 5

&nbsp;&nbsp;&nbsp;&nbsp;\*<br>
&nbsp;&nbsp;&nbsp;\*\*\* <br>
&nbsp;&nbsp;\*\*\*\*\* <br>
&nbsp;\*\*\*\*\*\*\* <br>
\*\*\*\*\*\*\*\*\* <br>

When **n** = 1 <br>

*

```python
n = 11

#This question DOES NOT requires you to do double (nested) for loop. 
#Understanding of the question is ALWAYS more important that jumping into it and randomly try

#If you notice, n actually represents the number of rows of the chrismas tree.
#So for the first for loop (usually represents the row), you can simply loop to the range of n
#and within each line is a little more tricky
#but if we look at the last line of each example. 

#   n       number of * at last line     how to get number of * at last line?   formula for num of *
#   3                  5                          (3*2) - 1 = 5                      (n*2) - 1
#   5                  9                          (5*2) - 1 = 9                      (n*2) - 1
#   1                  1                          (1*2) - 1 = 1                      (n*2) - 1

#with the understanding of this, we would know what are the total column each n will be.
#next all you need to do is to identify each row needs to add how many space and how many *
# so if we look at the n= 5 example (since that has the most sample)

# row num   num of space before *    num of *    how to get num of space    formula for space
#    1               4                  1               5 - 1 = 4               n-row_num
#    2               3                  3               5 - 2 = 3               n-row_num
#    3               2                  5               5 - 3 = 2               n-row_num
#    4               1                  7               5 - 4 = 1               n-row_num
#    5               0                  9               5 - 5 = 0               n-row_num

### num of * is calculated using the first formula we found above

#With both of these formuals, we can start to code!

#We don't want to start from 0, we wanna start from 1 this time
for i in range(1,n+1):
        
    row_num = i
    num_of_stars = (row_num * 2) - 1
    num_of_space = (n-row_num)

    #Recap, Strings can be added and multiplied using + and *
    # + means concatentation in this situation, when 2 strings are place with + then join together
    # * means multiply the number of times the string
    text_to_print_this_line = (" " * num_of_space) +  ("*" * num_of_stars)

    #Print the current line
    print(text_to_print_this_line)
```
