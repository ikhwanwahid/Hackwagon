### `get_all_latitudes()` - Functions to get all latitudes given a list of listing_ids <font color='red'>(5 marks)</font>


<i> Hint
1. Create a results list
2. Extract the latitude and listing id of each row
3. Check listing id exists within the all listings 
4. If true, append the latitude into the results list
5. Return results list
</i>

<br>

<b>Input</b>: `airbnb_data` as `data`, a `list` of listing_ids

<b>Return</b>: A `list` of latitudes


````python
#Write code below
# in the function, create an empty list
#extract out the latitude and the listing id
# if listing is part of what we input, append latitude
def get_all_latitudes(data,listing_id):    #two inputs , x input is where u get data from, 2nd input is listing we want
    results_list =[]
    for row in data:
        listing = row['listing_id']
        latitude= row['latitude']
        if listing in listing_id:
            results_list.append(latitude)
    return results_list
````

### `listings_recommender()` - Function to recommend all listings based on a given price, satisfaction score and neighborhood <font color='red'>(15 marks)</font>

Note:
1. Lesser than or equal to that price
2. Equal or more than that overall satisfaction score
3. In that neighborhood

<i> Hint
1. Create a results list
2. Extract the relevant of each row
3. Check it satisfies all conditions passed into the function
4. If true, append the listing id into the results list
5. Return results list
</i>

<b>Input</b>: `airbnb_data` as `data`, price, overall_satisfaction, neighborhood_id

<b>Return</b>: A `list` of listing_ids

````python
#Write code below
# input will be in following order x : airbnb_data, price,satisfaction_score,neighbourhood
# in the function, create an empty list
#extract out price,satisfaction_score,neighbourhood
#iterate if satisfy 3 conditions, append to empty list

def listings_recommender(data,input_price,input_satisfaction_score,input_neighborhood):
    results_list=[]
    for row in data:
        listing=row['listing_id']
        price=row['price']
        score=row['overall_satisfaction']
        neighborhood=row['neighborhood']
        if price <= input_price:
            if score >=input_satisfaction_score:
                if neighborhood==input_neighborhood:
                    results_list.append(listing)
    return results_list

        
````

### `price_recommender()` - Function to recommend a price in a neighborhood based on average price and overall satisfaction <font color='red'>(15 marks)</font>

For this function, we want to build a <b><i>simple</i></b> price recommendation function that will give a potential host a suggested price.

To build this, these are the requirements: 
1. Take all listings in that neighborhood and check for listings with a least 1 review and an overall satisfaction score of 4 or more.
2. From that filtered listings, calculate the average price and return that as the suggested price rounded to 2 decimal places.

<b>Input</b>: `airbnb_data` as `data`, a neighborhood_id

<b>Return</b>: A `float` of recommended price

````python
#Write code below
# input will be in following order x : airbnb_data,neighbourhood id
# in the function, create an empty list ....tis on dont need
# might need to add an empty library also just to contain  average price
# based on following conditions : all listings must have at least one review
#overall customer satisfaction >= 4
#calculate average price round =2
#output should be average price
def price_recommender(data,neighbourhood_choosen):
    results_list={}
    all_count={}
    for row in airbnb_data:
        neighbourhood_id=row['neighborhood']
        price= row['price']
        review =row['reviews']
        satisfaction=row['overall_satisfaction']
        if review >=1 and satisfaction>= 4:
            if neighbourhood_id not in results_list:
                results_list[neighbourhood_id]=price
                all_count[neighbourhood_id]=1
            else:
                results_list[neighbourhood_id]+=price
                all_count[neighbourhood_id]+=1
    average_price={}
    for i,j in results_list.items():
        id=i
        price = j
        count= all_count[i]
        average=j/count
        if id not in average_price:
            average_price[id]=average
        else:
            average_price[id]+=average
    return round(average_price[neighbourhood_choosen],2)





              
    
````
